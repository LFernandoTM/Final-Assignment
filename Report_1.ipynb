```python
import pandas as pd
%load_ext google.colab.data_table
df1 = pd.read_csv ("co-emissions-per-capita.csv")
df2 = pd.read_csv ("per-capita-energy-use.csv")
df3 = pd.read_csv ("gdp-per-capita-worldbank.csv")
df4 = pd.read_csv ("world-banks-income-groups.csv")


df1
#df2
#df3
#df4
```

```python
#merge CO2 emissions dataframe + Energy consumption per capita dataframe
df5 = df1.merge(df2, how = 'outer')
df5
```

```python
#merge CO2 emissions/Energy consumption per capita dataframe + GDP dataframe
df5 = df5.merge(df3, how = 'outer')
df5
```

```python
#merge CO2 emissions/Energy consumption per capita/GDP dataframe + Income classification
df5 = df5.merge(df4, how = 'outer')
df5
```

```python
#Check columns names
df5.columns.values
```

```python
#display list of countries
df5["Entity"].unique()
```

```python
#Entity as Index
df6 = df5.set_index("Entity")
df6
```

```python
#Delete innecessary values
df7 = df6.drop(["World","Upper-middle-income countries","South America","Oceania","North America (excl. USA)","North America","Lower-middle-income countries","Low-income countries","High-income countries","European Union (28)","European Union (27)","Europe (excl. EU-28)","Europe (excl. EU-27)","Europe","Asia (excl. China and India)","Asia","Africa"],axis=0)
df7
```

```python
#Delete Column "Code"
df8 = df7.drop(["Code"], axis=1)
df8
```

```python
#Rename columns
new_columns_names= ['Year', 'CO₂ emissions per capita (tonnes)', 'Energy consumption per capita (kilowatt-hours)','GDP per capita ($)','Income classification']
df8.columns=new_columns_names
display(df8.columns)
df8
```

```python
#Filter 2021 values
df_2021 = df8[df8["Year"]==2021]
df_2021
```

```python
#Delete NaN values
df_2021_clean = df_2021.dropna()
df_2021_clean
```

```python
#Graph
import matplotlib.pyplot as plt
df_2021_clean.plot(kind="scatter", x="CO₂ emissions per capita (tonnes)", y="Energy consumption per capita (kilowatt-hours)", c="GDP per capita ($)", cmap="magma")
plt.title('CO2 emissions per capita vs Energy consumption per capita vs GDP per capita (2021)')
plt.show()
```

```python
#sort by 2021 values
countries_co2_filter = df_2021_clean.sort_values(by="CO₂ emissions per capita (tonnes)", ascending=False)
#check top 20 countries with higher co2 emissions per capita
top_20 = countries_co2_filter.head(20)
top_20
```

```python
#Graph 2
import matplotlib.pyplot as plt
count_income_class= top_20['Income classification'].value_counts()
Income_classification = top_20['Income classification'].unique()
fig, ax = plt.subplots(figsize=(8,6))
plt.pie(count_income_class,labels=Income_classification,autopct='%0.0f %%',shadow=True)
ax.set_title("Income relation: Top 20 Countries biggest emission CO2 per capita")
ax.legend()
plt.show()
```
