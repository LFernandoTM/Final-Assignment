```python
import pandas as pd
%load_ext google.colab.data_table
df = pd.read_csv ("levelized-cost-of-energy.csv")
df
```

```python
#Delete NaN values
df1 = df.dropna()
df1
```

```python
#Delete "Code" column
df2 = df1.drop(["Code"], axis=1)
df2
```

```python
#Rename columns
new_columns_names= ['Country','Year', 'Bioenergy', 'Geothermal','Offshore wind','Solar photovoltaic','Solar power','Hidropower','Onshore wind']
df2.columns=new_columns_names
#display(df2.columns)
df2
```

```python
#Delete "Country" column
df3 = df2.drop(["Country"], axis=1)
df3
```

```python
#Reset index
df3 = df3.set_index("Year")
df3
```

```python
#Transpose rows and columns
df3_switched = df3.T
df3_switched
```

```python
#Reset index
df3_switched.reset_index(inplace = True)
df3_switched
```

```python
#Rename columns
df3_columns_names= ['Energy','2010', '2012', '2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']
df3_switched.columns=df3_columns_names
#display(df3.columns)
df3_switched
```

```python
#Graph 1
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
years = df3_switched.columns[1:]
for Energy, row in df3_switched.iterrows():
  plt.plot(years, row[1:], label = row[0])

plt.title('Levelized cost of energy (2010-2022)')
plt.xlabel("Years")
plt.ylabel("Energy cost ($)")
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.grid()
plt.show()
```

```python
#Predicted model and graph
import numpy as np
from sklearn.linear_model import LinearRegression

# Crear un DataFrame para almacenar las predicciones
predictions = pd.DataFrame(columns=['Year'])

# Generar los aÃ±os futuros (2023-2033)
future_years = np.arange(2023, 2034)
predictions['Year'] = future_years
years = df3_switched.columns[2:]

# Iterar sobre cada tipo de energÃ­a renovable
for index, row in df3_switched.iterrows():
    renewable_type = row['Energy']
    cost_data = row[2:]  # Datos de costos de energÃ­a
    cost_data = cost_data.values.reshape(-1, 1)  # Reshape para que sklearn pueda trabajar con ellos

    # Inicializar y ajustar el modelo de regresiÃ³n lineal
    model = LinearRegression()
    model.fit(np.array(years).reshape(-1, 1), cost_data)

    # Predecir los costos para los aÃ±os futuros
    future_predictions = model.predict(np.array(future_years).reshape(-1, 1))

    # Agregar las predicciones al DataFrame
    predictions[renewable_type] = future_predictions.flatten()

# GrÃ¡fica de las predicciones
plt.figure(figsize=(10, 6))
for renewable_type in df3_switched['Energy']:
    plt.plot(predictions['Year'], predictions[renewable_type], label=renewable_type)

plt.title('Predicted Levelized Cost of Renewable Energy (2023-2033)')
plt.xlabel("Years")
plt.ylabel("Energy cost ($)")
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()
plt.show()
```
