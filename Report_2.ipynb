```python
import pandas as pd
%load_ext google.colab.data_table
df = pd.read_csv ("annual-co2-emissions-per-country.csv")

df
```

```python
#Display list of countries
df["Entity"].unique()
```

```python
#Delete Column "Code"
df1 = df.drop(["Code"], axis=1)
df1
```

```python
#Filter by 2000
df_2000 = df1[df1["Year"]==2000]
df_2000
```

```python
#Filter by 2007
df_2007 = df1[df1["Year"]==2007]
df_2007
```

```python
#Filter by 2015
df_2015 = df1[df1["Year"]==2015]
df_2015
```

```python
#Filter by 2022
df_2022 = df1[df1["Year"]==2022]
df_2022
```

```python
#Rename columns
new_columns_names_2000= ['Entity', 'Year2000', 'Annual CO2 emissions 2000']
df_2000.columns=new_columns_names_2000
new_columns_names_2007= ['Entity', 'Year2007', 'Annual CO2 emissions 2007']
df_2007.columns=new_columns_names_2007
new_columns_names_2015= ['Entity', 'Year2015', 'Annual CO2 emissions 2015']
df_2015.columns=new_columns_names_2015
new_columns_names_2022= ['Entity', 'Year2022', 'Annual CO2 emissions 2022']
df_2022.columns=new_columns_names_2022
#df_2007
#df_2015
#df_2022
```

```python
#Merge 2000 + 2007
df2 = df_2000.merge(df_2007, how = 'outer')
df2
```

```python
#Merge 2000/2007 + 2015
df3 = df2.merge(df_2015, how = 'outer')
df3
```

```python
#Merge 2000/2007/2015 + 2022
df4 = df3.merge(df_2022, how = 'outer')
df4
```

```python
#Delete columns years
df5 = df4.drop(["Year2000","Year2007","Year2015","Year2022"], axis=1)
df5
```

```python
# add columns with increasing/decreasing co2 % per year since 2000

df5['2000'] = 100
df5['2007'] = ((df5['Annual CO2 emissions 2007'] / df5['Annual CO2 emissions 2000']) * 100)
df5['2015'] = ((df5['Annual CO2 emissions 2015'] / df5['Annual CO2 emissions 2000']) * 100)
df5['2022'] = ((df5['Annual CO2 emissions 2022'] / df5['Annual CO2 emissions 2000']) * 100)
df5
```

```python
#Column "Entity" replace index
df6 = df5.set_index("Entity")
df6
```

```python
#Delete innecessary values from "Entity"
df7 = df6.drop(["Africa","Africa (GCP)","Asia","Asia (GCP)", "Asia (excl. China and India)","Central America (GCP)","Europe (GCP)","European Union (28)","European Union (27)","Europe (excl. EU-28)","Europe (excl. EU-27)","Europe","International aviation","Middle East (GCP)","Non-OECD (GCP)","North America","North America (GCP)","North America (excl. USA)","OECD (GCP)","Oceania","Oceania (GCP)","World","Upper-middle-income countries","South America","Lower-middle-income countries","Low-income countries","High-income countries"],axis=0)
df7
```

```python
#Reset index
df7.reset_index(inplace = True)
df7
```

```python
#Delete NaN values
df8 = df7.dropna()
df8
```

```python
#Filter 2022 values
df8 = df8.sort_values(by='2022', ascending=True)
top_3 = df8.head(3)
top_3
```

```python
#Graph
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
years = top_3.columns[5:]
for Entity, row in top_3.iterrows():
  plt.plot(years, row[5:], label = row[0])

plt.title('Top 3 Countries biggest CO2 % decreased (2000-2022)')
plt.xlabel("Years")
plt.ylabel("CO2 Emissions (%)")
plt.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.grid()
plt.show()
```
